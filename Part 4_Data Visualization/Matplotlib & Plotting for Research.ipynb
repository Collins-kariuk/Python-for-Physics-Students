# Matplotlib

Matplotlib is a popular data visualization library in Python that provides a wide range of plotting options. It is highly customizable and can be used to create a variety of plots, including line plots, bar plots, scatter plots, histograms, etc.

## Basics of Plotting Data

Matplotlib provides two interfaces for plotting data:

- the `pyplot` interface and
- the object-oriented interface.

The `pyplot` interface is a high-level interface that is easy to use and is suitable for simple plots. The object-oriented interface is more flexible and allows for more advanced customization.

In this notebook, we will stick to the `pyplot` interface to create various types of plots.

Here are some common plot types:

### Line Plot

Line plots are a great way to visualize data points in a sequence, highlighting the relationship between them. They are particularly useful for showing trends over time or comparing multiple datasets. A line plot is created by connecting data points with straight lines. In Physics, a common example might be plotting the position of an object over time to analyze its motion.

````python
# Line plot example: Position vs Time in Physics

import matplotlib.pyplot as plt

# Create some data: time (seconds) and position (meters)
time = [0, 1, 2, 3, 4, 5]  # Time in seconds
position = [0, 5, 20, 45, 80, 125]  # Position in meters (e.g., free-falling object with increasing velocity)

# Create a line plot
plt.plot(time, position, marker='o', linestyle='-', color='blue')

# Add a title
plt.title('Position vs Time')

# Add labels to the axes
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')

# Show the plot
plt.show()

# Save the plot to a file (optional)
# plt.savefig('position_vs_time.png')
````

In the example above, you can customize the line plot by adjusting its color, line style, and marker style. For example, to change the line color to red, set the `color` parameter to `'r'`. To make the line dashed, use the `linestyle` parameter with `'--'`, and to set the marker style to circles, specify `marker='o'`. Experimenting with these parameters will help you understand how they affect the appearance of the plot. For more customization options, check the [Matplotlib Line Plot Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) or watch this YouTube video: [Introduction to Line Plot Graphs with Matplotlib Python](https://www.youtube.com/watch?v=AYorFcI1MTU).

### Scatter Plot

Scatter plots are an excellent tool for visualizing the relationship between two variables and are particularly useful for identifying patterns, trends, and outliers in the data. For example, you could use a scatter plot to examine the relationship between a student's study time and their exam score. In Physics, scatter plots are often used to visualize relationships such as the correlation between force and acceleration in an experiment.

````python
# Scatter plot example:

import matplotlib.pyplot as plt

# Experimental data
mass = 5  # in kilograms
acceleration = [1, 2, 3, 4, 5, 6]  # in m/s^2
force = [m * mass for m in acceleration]  # Force = mass * acceleration

# Create a scatter plot
plt.scatter(acceleration, force, color='blue', marker='o')

# Add titles and labels
plt.title("Force vs. Acceleration")
plt.xlabel("Acceleration (m/s²)")
plt.ylabel("Force (N)")

# Show the plot
plt.grid(True)
plt.show()
````

💡 Notice how we calculated force above using `force = [m * mass for m in acceleration]`. This approach is an example of **list comprehension**. Simply put, list comprehension is a concise way to create lists by applying an operation to each item in an iterable (like a list or range) within a single line of code. It’s a clean and Pythonic way to write compact, readable code.

### Bar Charts

Bar charts are a widely used method for visualizing categorical data, allowing you to compare the values of different categories easily. They are especially useful for representing non-continuous or discrete data. For instance, you might use a bar chart to compare the sales of various products or the outcomes of different experiments.

````python
# Bar plot example:

import matplotlib.pyplot as plt

experiments = ["Exp 1", "Exp 2", "Exp 3"]
results = [92, 85, 78]

plt.bar(experiments, results, color="blue")
plt.title("Experiment Results")
plt.xlabel("Experiment")
plt.ylabel("Score (%)")

plt.show()
````

## Customizing Plots

Customizing plots is useful for creating visually appealing and informative visualizations. Matplotlib provides a wide range of customization options, allowing you to control the appearance of your plots in detail.

__Titles, Labels, and Legends__:

- You can add a title to your plot using the `plt.title()` function.
- You can label the x-axis and y-axis using the `plt.xlabel()` and `plt.ylabel()` functions, respectively.
- You can include a legend in your plot using the `plt.legend()` function. Legends are essential for labeling different elements of a plot, helping viewers easily identify and understand the data being visualized.
-  You can customize the font size, font weight, font style, and font color of the title, labels, and legend using the `fontsize`, `fontweight`, `fontstyle`, and `color` parameters.

__Colors, Styles, and Markers__:

- You can change the color of the plot using the `color` parameter.
- You can change the line style using the `linestyle` parameter. Common line styles include solid lines (`'-'`), dashed lines (`'--'`), and dotted lines (`':'`).
- You can change the marker style using the `marker` parameter. Common marker styles include circles (`'o'`), squares (`'s'`), and triangles (`'^'`).

__Adjusting Plot Size__:

- You can adjust the size of the plot using the `plt.figure()` function. The `figsize` parameter allows you to set the width and height of the plot in inches. For example, `plt.figure(figsize=(10, 6))` creates a plot with a width of 10 inches and a height of 6 inches.

## Example: Plotting Time-Series Data

Suppose you’re analyzing data from an experiment tracking an object’s motion over time:

````python
import pandas as pd

data = {
    "Time (s)": [0, 1, 2, 3, 4, 5],
    "Position (m)": [0, 5, 20, 45, 80, 125],
    "Velocity (m/s)": [0, 5, 10, 15, 20, 25],
}

df = pd.DataFrame(data)

# Plotting Position and Velocity vs Time
plt.plot(df["Time (s)"], df["Position (m)"], label="Position")
plt.plot(df["Time (s)"], df["Velocity (m/s)"], label="Velocity")

plt.title("Position and Velocity vs Time")

plt.xlabel("Time (s)")
plt.ylabel("Measurement")

plt.legend()

plt.grid(True)
plt.show()
````

# Plotting for Research

## Best Practices for Data Visualization

When creating data visualizations for research purposes, it’s essential to follow best practices to ensure that your plots are clear, accurate, and informative. Here are some key tips to keep in mind:

__1. Choose the Right Plot Type__: Select a plot type that effectively represents the data and highlights the relationships you want to convey. For example, use line plots for time-series data, scatter plots for examining relationships between variables, and bar charts for comparing categories.

__2. Label Axes and Add Legends__: Always label the x-axis and y-axis of your plots to provide context and clarity. If your plot includes multiple datasets, add a legend to distinguish between them.

__3. Use Color Wisely__: Color can be a powerful tool for highlighting information in your plots. However, avoid using too many colors, as this can make the plot difficult to interpret. Choose a color palette that is easy to read and consider using color-blind-friendly palettes for accessibility.

__4. Provide Context with Titles and Annotations__: Add a title to your plot to provide an overview of the data being visualized. You can also include annotations or text labels to highlight specific points of interest in the plot.

__5. Customize Your Plot__: Experiment with customizing your plots to make them visually appealing and informative. Adjust the plot size, font styles, line styles, and markers to enhance the readability of your visualizations.

__6. Keep It Simple__: Avoid cluttering your plots with unnecessary elements. Focus on conveying the key insights from your data in a clear and concise manner.

# Exporting Plots in Various Formats

Once you have created a plot, you may want to export it in different formats for use in reports, presentations, or publications. Matplotlib allows you to save plots in various file formats, such as PNG, JPEG, PDF, and SVG.

To save a plot in a specific format, you can use the `plt.savefig()` function and specify the file name along with the desired file format. For example:

```python
# Save the plot
plt.savefig("position_vs_time.png", dpi=300) # Save as PNG with high resolution

```

### Simulating Projectile Motion

````python
# Simulate projectile motion

import numpy as np # Import NumPy library (don't worry about this for now, we'll cover NumPy later)

time = np.linspace(0, 5, 100)
g = 9.8
v0 = 20

angle = np.radians(45)
x = v0 * np.cos(angle) * time

y = v0 * np.sin(angle) * time - 0.5 * g * time**2

plt.plot(x, y, label="Projectile Path", color="green")
plt.title("Projectile Motion")
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")
plt.legend()
plt.grid(True)

plt.show()
````

### Comparing Experimental and Theoretical Results

````python
# Experimental data

experimental = [0, 4.8, 19.6, 44.1, 78.4]
theoretical = [0, 5, 20, 45, 80]
time = [0, 1, 2, 3, 4]

plt.plot(time, experimental, label="Experimental", linestyle="--", marker="o")
plt.plot(time, theoretical, label="Theoretical", linestyle="-", marker="s")

plt.title("Experimental vs Theoretical Results")

plt.xlabel("Time (s)")
plt.ylabel("Position (m)")

plt.legend()
plt.grid(True)
plt.show()
````

### Gaussian Distribution Visualization

````python
import numpy as np

# Generate data
x = np.linspace(-3, 3, 100)
y = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

# Plot

plt.plot(x, y, label="Gaussian Curve", color="blue")
plt.title("Visualization of a Gaussian Distribution")
plt.xlabel("X-axis")
plt.ylabel("Probability Density")

plt.legend()
plt.grid(True)
plt.show()
````
