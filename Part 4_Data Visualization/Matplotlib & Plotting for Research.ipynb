# Matplotlib

Matplotlib is a popular data visualization library in Python that provides a wide range of plotting options. It is highly customizable and can be used to create a variety of plots, including line plots, bar plots, scatter plots, histograms, etc.

## Basics of Plotting Data

Matplotlib provides two interfaces for plotting data:

- the `pyplot` interface and
- the object-oriented interface.

The `pyplot` interface is a high-level interface that is easy to use and is suitable for simple plots. The object-oriented interface is more flexible and allows for more advanced customization.

In this notebook, we will stick to the `pyplot` interface to create various types of plots.

Here are some common plot types:

### Line Plot

Line plots are a great way to visualize data points in a sequence, highlighting the relationship between them. They are particularly useful for showing trends over time or comparing multiple datasets. A line plot is created by connecting data points with straight lines. In Physics, a common example might be plotting the position of an object over time to analyze its motion.

```python
# Line plot example: Position vs Time in Physics

import matplotlib.pyplot as plt

# Create some data: time (seconds) and position (meters)
time = [0, 1, 2, 3, 4, 5]  # Time in seconds
position = [0, 5, 20, 45, 80, 125]  # Position in meters (e.g., free-falling object with increasing velocity)

# Create a line plot
plt.plot(time, position, marker='o', linestyle='-', color='blue')

# Add a title
plt.title('Position vs Time')

# Add labels to the axes
plt.xlabel('Time (s)')
plt.ylabel('Position (m)')

# Show the plot
plt.show()

# Save the plot to a file (optional)
# plt.savefig('position_vs_time.png')
```

In the example above, you can customize the line plot by adjusting its color, line style, and marker style. For example, to change the line color to red, set the `color` parameter to `'r'`. To make the line dashed, use the `linestyle` parameter with `'--'`, and to set the marker style to circles, specify `marker='o'`. Experimenting with these parameters will help you understand how they affect the appearance of the plot. For more customization options, check the [Matplotlib Line Plot Documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html) or watch this YouTube video: [Introduction to Line Plot Graphs with Matplotlib Python](https://www.youtube.com/watch?v=AYorFcI1MTU).

### Scatter Plot

Scatter plots are an excellent tool for visualizing the relationship between two variables and are particularly useful for identifying patterns, trends, and outliers in the data. For example, you could use a scatter plot to examine the relationship between a student's study time and their exam score. In Physics, scatter plots are often used to visualize relationships such as the correlation between force and acceleration in an experiment.

```python
# Scatter plot example:

import matplotlib.pyplot as plt

# Experimental data
mass = 5  # in kilograms
acceleration = [1, 2, 3, 4, 5, 6]  # in m/s^2
force = [m * mass for m in acceleration]  # Force = mass * acceleration

# Create a scatter plot
plt.scatter(acceleration, force, color='blue', marker='o')

# Add titles and labels
plt.title("Force vs. Acceleration")
plt.xlabel("Acceleration (m/s²)")
plt.ylabel("Force (N)")

# Show the plot
plt.grid(True)
plt.show()
```

💡 Notice how we calculated force above using `force = [m * mass for m in acceleration]`. This approach is an example of **list comprehension**. Simply put, list comprehension is a concise way to create lists by applying an operation to each item in an iterable (like a list or range) within a single line of code. It’s a clean and Pythonic way to write compact, readable code.

### Bar Charts

Bar charts are a widely used method for visualizing categorical data, allowing you to compare the values of different categories easily. They are especially useful for representing non-continuous or discrete data. For instance, you might use a bar chart to compare the sales of various products or the outcomes of different experiments.

```python
# Bar plot example:

import matplotlib.pyplot as plt

experiments = ["Exp 1", "Exp 2", "Exp 3"]
results = [92, 85, 78]

plt.bar(experiments, results, color="blue")
plt.title("Experiment Results")
plt.xlabel("Experiment")
plt.ylabel("Score (%)")

plt.show()
```
